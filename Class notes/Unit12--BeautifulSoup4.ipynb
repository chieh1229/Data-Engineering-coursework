        - This attribute checks if the status code of the response is between 400 and 600 to see if there was a client error or a server error.

    - raise_for_status() : Raises stored HTTPError, if one occurred.
    - <font color=#0000FF>reason</font> = None : Textual reason of responded HTTP Status, e.g. “Not Found” or “OK”.
    - request = None : The PreparedRequest object to which this is a response.
    - <font color=#0000FF>status_code</font> = None : Integer Code of responded HTTP Status, e.g. 404 or 200.

```python
url = "https://www.chinatimes.com"
resp = requests.request(method="GET", url=url)
soup = BeautifulSoup(markup=resp.text, features="html.parser")
nodes = soup.select(".col-md-6 .col .title a")

# print([node.text for node in nodes])
# print("\n----------------------------------------------\n")
# # print([node.get("href") for node in nodes])

titles = [node.text for node in nodes]
hrefs = [node.get('href') for node in nodes]
```

```python
contents = []
time = []

for href in hrefs :
    resp = requests.request(method='GET', url=href)
    soup = BeautifulSoup(markup=resp.text, features='html.parser')
    
    nodes = soup.select('div.article-body > p')
    contents.append(''.join([node.text.replace("\n", "") for node in nodes]))
 
    nodes = soup.select('div.col-md > div > div > time')
    time.append(''.join([node.text for node in nodes]))

df = pd.DataFrame({'新聞標題':titles,
                   '超連結':hrefs,
                   '內文':contents,
                   '發布時間':time})
df
```

```python
url = "https://udn.com/rank/pv/2"
resp = requests.request(method="GET", url=url)
soup = BeautifulSoup(markup=resp.text, features="html.parser")
nodes = soup.select("div.story-list__text > h2 > a")

titles = [node.text for node in nodes]
hrefs = [node.get('href') for node in nodes]

contents = []
time = []

for href in hrefs :
    resp = requests.request(method='GET', url=href)
    soup = BeautifulSoup(markup=resp.text, features='html.parser')

    nodes = soup.select('div.story-list__text > p')
    contents.append(''.join([node.text.replace("\n", "") for node in nodes]))

    nodes = soup.select('div.article-content__subinfo > section.authors > time')
    time.append(''.join([node.text for node in nodes]))

df = pd.DataFrame({'新聞標題':titles,
                   '超連結':hrefs,
                   '內文':contents,
                   '發布時間':time})
df
```

    1. resp = requests.get(url)
    2. soup = resp.text
    3. nodes = soup.select(selector="form")        #找出填寫資料的 form，並找出變數名稱與對應的value
    4. formData = {key1:value1, key2:value2, ...}  #此處的key1, key2, ...就是form內的變數名稱
    5. sess = requests.session()                   #建立一個連線的 session
    6. result = sess.post(url=?, data=formData)    #找出執行程式的url，並將 data 餵給它，會應回來的又是一個HTML
    7. 重新 parse 6. 回傳的 HTML

### <font color=red>有login的動作，一定要先建立 session (物件)，再透過此 session 進行 request</font>
- 【使用時機】該網站需要 login 後，才能繼續爬文時
- 【步驟】
    1. 建立 session 物件
    2. 透過此 session 物件進行 request(s)
    3. 爬取 response 物件內的資料

#### <font color=blue>state-less and state-ful</font>
<img src="http://180.218.162.136:2018/Images/Spider/session.jpg" width=500 align=left></img>

```python
url = "https://www.drmaster.com.tw/Publish_Search.asp"
payloads = {"selectID": "3",
            "content": "陳祥輝"}

# step 1 : 建立session物件
sess = requests.session()

# step 2 : 基於此 sess 進行 request
resp = sess.request(method="POST", url=url, data=payloads)

resp.encoding="utf-8"
result = pd.read_html(resp.text, header=0)
result[9]
```

#### 設定 cookie 後，就會直接進入 ( 在Headers中找到 set-cookie : over18 = 1 (是); path = / )

```python
url = "https://www.ptt.cc/bbs/Gossiping/index.html"
cookies = {"over18":"1"}
resp = requests.request(method="POST", url=url, cookies=cookies)
print(resp.text)
```

```python
soup = BeautifulSoup(markup=resp.text, features="html.parser")
nodes = soup.select("div.title > a")
titles = [node.text for node in nodes]
hrefs = [node.get('href') for node in nodes]
nodes = soup.select("div.meta > div.date")
times = [node.text for node in nodes]

df = pd.DataFrame({'標題':titles,
                   '超連結':hrefs,
                   '發布時間':times})
df
```
