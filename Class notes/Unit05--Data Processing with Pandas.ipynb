Pandas 提供的重要資料結構
1. Series：用來處理時間序列相關的資料(如感測器資料等)，主要為建立索引的一維陣列。
2. DataFrame：用來處理結構化(Table like)的資料，有列索引與欄標籤的二維資料集，例如關聯式資料庫、CSV 等等。

###### DataFrame 的 concat

```python
# (type 1 data) 欄位數、名稱一樣，適合上下/左右合併
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']})

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']})
```

```python
print(df1)
print(df2)
```

```python
pd.concat([df1, df2], axis=0, verify_integrity=True)  # verify_integrity=True驗證資料，index有重複所以error
```

#### <font color=red>【非常重要】上、下合併，axis=0</font>

```python
pd.concat([df1, df2], axis=0, verify_integrity=True, ignore_index=True)
```

```python
# (type 2 data : column名稱不同)
df3 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']})

df4 = pd.DataFrame({'E': ['A4', 'A5', 'A6', 'A7'],
                    'F': ['B4', 'B5', 'B6', 'B7'],
                    'G': ['C4', 'C5', 'C6', 'C7'],
                    'H': ['D4', 'D5', 'D6', 'D7']})
```

```python
print(df3)
print(df4)
```

#### <font color=red>【非常重要】左、右合併，axis=1</font>

```python
pd.concat([df3, df4], axis=1, verify_integrity=True)    # 左右合併，如果ignore_index欄位名稱會不見
```
